```python
import random

randNums = []

for m in range(1, 1001):
    numbers = random.randint(1, 100000)
    randNums.append(numbers)


print("Random numbers :", randNums)
print("Length :", len(randNums))

summ = sum(randNums)
print("Sum :", summ)
m = summ/(len(randNums))
print("Mean :", m) 


difference=[]
for a in randNums:
    difference.append((a-m)**2)
    

d=sum(difference)

sd=(d/(len(randNums)-1)**0.5)

z_score=[]
for z in randNums:
    z_score.append((z-m)/sd)
outlier=[]
for i in z_score:
    if (i>2):
        outlier.append(i)
print(outlier)
   
```

```python
import random


randNums = []

for m in range(1, 11):
    numbers = random.randint(1, 100)
    randNums.append(numbers)

"""  Normalization """
for m in randNums:
    l =( m-min(randNums))/(max(randNums)-min(randNums))
    print("Number :", m, "Normalization :", l)
   
""" Standardization"""
summ = 0
for i in randNums:
    summ += i

mean = summ/len(randNums)
print("Mean :", mean)

cig = 0
s = 0
for k in randNums:
    cig = (k-mean)**2
    s += cig

print("Sum :", s)
r = (s/len(randNums))**0.5
print("Square root :", r)
for o in randNums:
    z = (o-mean)/r
    print("Number :", o, "Standardization :", abs(z))
    
    
    
```
